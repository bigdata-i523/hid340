```python
class Comparison:
    
    def __init__(self, file_pointer):
        self.file_pointer = file_pointer
    
    def _load_json(self):
        with open(self.file_pointer, "r") as fp:
            return json.load(fp)
    
    def process(self):
        for item in self._load_json()["records"]:
            yield item                 
            
    def _get_titles(self, source, target):
        title_test = {"titles": [strcmp.mod_editdist(source["title"], target["title"])],
                      "subtitles": [strcmp.mod_editdist(source["title"], target["title"] + \
                                                        target["subtitle"])]}
        if title_test["titles"] >= title_test["subtitles"]:
            return source["title"], target["title"]
        else:
            return source["title"], target["title"] + target["subtitle"]
    
    def compare_titles(self, source, target):
        source_title, target_title = self._get_titles(source, target)
        return compare_all(source_title, target_title)
        
    def compare_values(self, source, target):
        for item in source:           
            if item:
                for t in target:
                    if item == t:
                        return compare_all(item, t)
                    else:
                        return compare_all(item, t)
        return np.array([np.nan for i in range(8)])
        
    def compare_value(self, source, target):
        if source and target:
            return compare_all(source, target)
        else:
            return np.array([np.nan for i in range(8)])        
            
    def _get_entities(self, source, target):       
        s = []
        t = []
        if source:
            for sc in source:
                s.append(sc)
                for tc in target:
                    t.append(tc)       
        return s, t     
    
    def compare_entities(self, source, target):
        
        s, t = self._get_entities(source, target)
        for x in s:
            for y in t:
                if x and y:
                    return compare_all(x, y)             
        return np.array([np.nan for i in range(8)])
    
    
    
def compare_all(s, t):
    
    return np.array([strcmp.editdist(s, t), 
            strcmp.mod_editdist(s, t),
            strcmp.swdist(s, t),
            strcmp.bigram(s, t),
            strcmp.qgram(s, t, q=3),
            strcmp.posqgram(s, t),
            strcmp.lcs(s, t),
            strcmp.syllaligndist(s, t)])
#, np.float64(1.)  

comp = Comparison("records.json")
    
np.save("record_pairs.npy",
    np.array([    
        np.array([
          [date for date in comp.compare_value(source["date"], target["date"])],
          [isbn for isbn in comp.compare_values(source["isbn"], target["isbn"])],
          [title for title in comp.compare_titles(source, target)],
          [creator for creator in comp.compare_entities(source["creator"], target["creator"])],
          [contributor for contributor in comp.compare_entities(source["contributor"], target["contributor"])],          
          #[1. for bias in range(9)],
          [target["match"] for i in range(8)],          
          [source["key"] for i in range(8)]
        ])
        for source, target in comp.process()
    ])
)

                    

    
    
```
